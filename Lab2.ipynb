```python
from ipywidgets import widgets
from IPython.display import display
from IPython.display import display, clear_output

from ipywidgets import Button, Layout

def booths_algorithm(mcand, plier):
    print(f"{int(mcand,2)},{int(plier,2)} \n multipcand: {mcand.lstrip('0')} ; multiplier: {plier.lstrip('0')}")
    product = "00000000" + plier + "0"
    print("Product: " + product.lstrip('0') )
    for i in range(1,9):
        operation = product[len(product)-2:]
        product = perform_operation(product,mcand,operation)
    product = shift(product)
    product = product[9:17]
    print("Result: " + product.lstrip('0') ,
         "Int : ",int(product,2)
         )
    return

def perform_operation(product,mcand,operation):
    if operation == "00" or operation == "11" :
        product = shift(product)
        print("Nothing do, only shift")
        return product
    elif operation == "01":
        print(product[0:8], product[8:])
        temp = bin_add(product[0:8],mcand)
        product = temp + product[8:]
        print(f"Add => Product  + Multiplicand  = Product  \n")
        product = shift(product)
        return product
    
    elif operation == "10":
        product = bin_subtr(product,mcand)
        print(f"Subtraction => Product  - Multiplicand  = Product  \n")
        product = shift(product)
        
        return product
    else:
        print("Error")
        return 0


def bin_subtr(product,mcand):
    carry = 0
    prime_product = product[:8]
    final_product = ""
    for i in range(len(prime_product)-1,-1,-1):
        if (mcand[i] == "0" and prime_product[i] == "0"):
            if (carry == 1):
                final_product = "1" + final_product
            else:
                final_product = "0" + final_product
        elif (mcand[i] == "1" and prime_product[i] == "0"):
            if (carry == 1):
                final_product = "0" + final_product
            else:
                final_product = "1" + final_product
                carry = 1
        elif (mcand[i] == "0" and prime_product[i] == "1"):
            if (carry == 1):
                final_product = "0" + final_product
                carry = 0
            else:
                final_product = "1" + final_product
        elif (mcand[i] == "1" and prime_product[i] == "1"):
            if (carry == 1):
                final_product = "1" + final_product
                carry = 1
            else:
                final_product = "0" + final_product
        else:
            print("Error")
            return 0

    return final_product + product[8:]


def shift(product):
    product = "0"+product[:len(product)-1]
    return product

def flip(string):
    flipped_string = ""
    return ''.join([flipped_string.join('0') if i=='1' else '1' for i in string])



def bin_add(num1, num2):
    num1 = num1.zfill(max(len(num1), len(num2))) 
    num2 = num2.zfill(max(len(num1), len(num2)))
    carry = 0
    result = ""
    for i in range(len(num1)-1, -1, -1): 
        cur_sum = int(num1[i]) + int(num2[i]) + carry
        if cur_sum == 0:
            carry = 0
            result = "0" + result
        elif cur_sum == 1:
            carry = 0
            result = "1" + result
        elif cur_sum == 2:
            carry = 1
            result = "0" + result
        elif cur_sum == 3:
            carry = 1
            result = "1" + result
    if carry == 1:
        result = "1" + result
    
    return result

def converts_negative_numbers(number):
    adjusted = abs(int(number))
    binint = "{0:b}".format(adjusted)
    binint=binint.zfill(8)
    flipped = flip(binint)
    r= bin_add(flipped,'1')
    return r

def convertDec(dec):
    
    if int(dec)<0:
        bin = converts_negative_numbers(int(dec))
   
    else:
        bin = "{0:b}".format(int(dec))
        
        for i in range(8-len(bin)):
            bin = "0" + bin
    print(type(bin),dec ,bin)
    return bin
```

```python
def binary_division(dividend, divisor):
    quotient = ""
    remainder = ""

    for i in range(len(dividend)):
        remainder += dividend[i]

        if int(remainder, 2) >= int(divisor, 2):
            quotient += "1"
            remainder = bin(int(remainder, 2) - int(divisor, 2))[2:]
        else:
            quotient += "0"
    quotient = quotient.lstrip('0')
    if quotient == "":
        quotient = "0"   
        
    print(f"Result quotient  {quotient.lstrip('0')}, remainder : {remainder}\n")
    quotient = bin(int(quotient, 2))[2:]
    remainder = bin(int(remainder, 2))[2:]
    print(f"Int Result quotient  {int(quotient,2)}, remainder : {int(remainder,2)}\n")
    return quotient, remainder
```

```python
import struct

def float_to_ieee754(x):
    x_bits = format(struct.unpack('!I', struct.pack('!f', x))[0], '032b')
    return x_bits

def ieee754(N): 
    a = int(N[0])        
    b = int(N[1:9],2) 
    c = int("1"+N[9:], 2)
    return (-1)**a * c /( 1<<( len(N)-9 - (b-127) ))

def ieee754_addition(X1, X2):
    S1, E1, M1 = X1[0], X1[1:9], X1[9:]
    S2, E2, M2 = X2[0], X2[1:9], X2[9:]
    E1 = int(E1, 2)
    E2 = int(E2, 2)

    if E1 < E2:
        X1, X2 = X2, X1
        S1, S2 = S2, S1
        M1, M2 = M2, M1
        E1, E2 = E2, E1
    E3 = E1
    Exp_diff = E1 - E2
    M2 = '1' + M2
    for i in range(Exp_diff):
        M2 += '0'
    if S1 == S2:
        M3 = bin(int(M1, 2) + int(M2, 2))[2:]
    else:
        M3 = bin(int(M1, 2) - int(M2, 2))[2:]
    while len(M3) > 24:  
        M3 = M3[:-1]
        E3 += 1
    if M3[0] == '0':
        M3 = M3[1:]
        E3 -= 1

    E3 = bin(E3)[2:].zfill(8)

    result = S1 + E3 + M3.zfill(23)

    return result

def ieee_754_main_function(num1,num2):
    x1= float_to_ieee754(num1)
    x2= float_to_ieee754(num2)
    print(f"""Float to IEEE-754 : \nx1 = {x1} \nx2 = {x2}"""
         )
    result =ieee754_addition(x1,x2)
    print(f'Result IEEE-754 format {result}')
    print(f'Convert to Float {ieee754(result)}')
    
    
    
```
